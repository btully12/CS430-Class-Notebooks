```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
np.random.seed(0)
X= 2 - 3 * np.random.normal(0,1,20)
y= X - 5 * (X**2) + 0.5 * (X**3) + np.random.normal(-3,3,20)
```

```python
# Increase the dimension
X= X[:, np.newaxis]
y=y[:, np.newaxis]
plt.scatter(X,y)
```

```python
lr=LinearRegression()
linear_model=lr.fit(X,y)
y_pred=linear_model.predict(X)
plt.scatter(X,y)
plt.plot(X,y_pred, color="red")
plt.xlabel("X")
plt.ylabel("y")
plt.show()
RMSE=mean_squared_error(y, y_pred, squared=False)
print(RMSE)
```

## Poly regression

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import Pipeline
```

```python
poly_model=Pipeline([('Polynomial', PolynomialFeatures(degree=3)),
                     ('Model', LinearRegression())])
```

```python
poly_model.fit(X,y)
```

```python
y_pred_poly=poly_model.predict(X)
```

```python
coeff=poly_model.named_steps['Model'].coef_
print(coeff)
```

```python
intercept=poly_model.named_steps['Model'].intercept_
print(intercept)
```

```python
RMSE=mean_squared_error(y, y_pred_poly, squared=False)
RMSE
```

```python
# Sorting predicted values 
sorted_zip= sorted(zip(X, y_pred_poly))
X_poly, y_pred_poly=zip(*sorted_zip)
```

```python
plt.scatter(X,y)
plt.plot(X_poly,y_pred_poly,color='blue', label='degree=10')
plt.xlabel("X-Poly")
plt.ylabel("y_poly")
plt.legend(loc='lower right')
plt.show()
```
