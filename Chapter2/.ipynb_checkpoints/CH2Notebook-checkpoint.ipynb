## Using the processed data

```python
house1 = pd.read_csv("house_processed.csv")
house1
```

```python
house1 = house1.drop(house1.columns[[0, 1, 2]],axis = 1)
```

# Feature Scaling

```python

from sklearn.preprocessing import MinMaxScaler, StandardScaler 
SC = StandardScaler()
MM=MinMaxScaler()
X = SC.fit_transform(house1)
X1 = MM.fit_transform(X)
print(X1)
```

# Train Test Split

```python
from sklearn.model_selection import train_test_split 
```

```python
training_set,test_set=train_test_split(X1,test_size=.5)
training_set
```
