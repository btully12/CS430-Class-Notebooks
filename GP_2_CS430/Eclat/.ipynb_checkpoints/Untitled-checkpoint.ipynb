```python
import numpy as np
import pandas as pd 
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
from pyECLAT import ECLAT
```

```python
df = pd.read_csv('marketb.csv')
```

```python
df
```

```python
transactions = []
for sublist in df.values.tolist():
    clean_sublist = [item for item in sublist if item is not np.nan]
    transactions.append(clean_sublist)
```

```python
transaction = pd.DataFrame(transactions) 
transaction
```

# Looking for item sets & do not want to have any individual products returned


```python
min_n_products = 2
```

# Want the min support to 7 but show it as percentage

```python
min_support = 7/len(transactions)
```

```python
max_length = max([len(x) for x in transactions])
```

# Creating model

```python
eclat = ECLAT(data=transaction, verbose=3)
```

# Fitting the model with the parameters above

```python
rule_indices, rule_supports = eclat.fit(min_support=min_support, 
                                        min_combination=min_n_products,
                                        max_combination=max_length)
```

```python
print(rule_supports)
```
